```python
#Now it's time for plotting the graph

#This is a funtion for adujsting the size of the graph. where 20 is the width size and 10 is the height of the graph 
plt.figure(figsize=(20,10))

#We're plotting individual bar for each employee
plt.bar(x[0],r[0],color='r')
plt.bar(x[1],r[1],color='#a252d1')
plt.bar(x[2],r[2],color='#48af8a')
plt.bar(x[3],r[3],color='y')

#Labeling the data...
plt.title('Average duty hours')
plt.xlabel('Employees')
plt.ylabel('Hour')
plt.show()
```
